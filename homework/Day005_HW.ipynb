```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
```

```python
import json

data=json.loads(r.text)
print('API ä¸€æ¬¡æœƒå›å‚³',len(data),'ç­†è³‡æ–™')
```

```python
r_dict=data[0]
print('æ¯ä¸€ç­†è³‡æ–™åŒ…å«ä¸‹åˆ—æ¬„ä½')

for i in r_dict:
    print(str(i))
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i in data:
    print('id',i['id'])
    print('æ¨™é¡Œ',i['title'])
    print('è²¼æ–‡æ™‚é–“',i['createdAt'])
    print('ç•™è¨€äººæ•¸',i['commentCount'])
    print('æŒ‰è®šäººæ•¸',i['likeCount'],'\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

# ç†±é–€
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
popular_data=json.loads(r.text)

pr_mean_comment, pr_mean_like, nopr_mean_comment, nopr_mean_like = [],[],[],[]

for i in popular_data:
    pr_mean_comment.append(i['commentCount'])
    pr_mean_like.append(i['likeCount'])

# éç†±é–€
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
r.text
nopopular_data=json.loads(r.text)

for i in nopopular_data:
    nopr_mean_comment.append(i['commentCount'])
    nopr_mean_like.append(i['likeCount'])
    

print("ç†±é–€æ–‡ç« :")
print("ç•™è¨€äººæ•¸:", sum(pr_mean_comment), " å¹³å‡ç•™è¨€äººæ•¸:", round(sum(pr_mean_comment)/float(len(popular_data)),0))
print("æŒ‰è®šäººæ•¸", sum(pr_mean_like), " å¹³å‡æŒ‰è®šäººæ•¸:", round(sum(pr_mean_like)/float(len(popular_data)),0),'\n')

print("éç†±é–€æ–‡ç« :")
print("ç•™è¨€äººæ•¸:", sum(nopr_mean_comment), " å¹³å‡ç•™è¨€äººæ•¸:", round(sum(nopr_mean_comment)/float(len(nopopular_data)),0))
print("æŒ‰è®šäººæ•¸", sum(nopr_mean_like), " å¹³å‡æŒ‰è®šäººæ•¸:", round(sum(nopr_mean_like)/float(len(nopopular_data)),0),'\n')
```
