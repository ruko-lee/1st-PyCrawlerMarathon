## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
```

```python
import json

data=json.loads(r.text)
print('API 一次會回傳',len(data),'筆資料')
```

```python
r_dict=data[0]
print('每一筆資料包含下列欄位')

for i in r_dict:
    print(str(i))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for i in data:
    print('id',i['id'])
    print('標題',i['title'])
    print('貼文時間',i['createdAt'])
    print('留言人數',i['commentCount'])
    print('按讚人數',i['likeCount'],'\n')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

# 熱門
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
popular_data=json.loads(r.text)

pr_mean_comment, pr_mean_like, nopr_mean_comment, nopr_mean_like = [],[],[],[]

for i in popular_data:
    pr_mean_comment.append(i['commentCount'])
    pr_mean_like.append(i['likeCount'])

# 非熱門
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
r.text
nopopular_data=json.loads(r.text)

for i in nopopular_data:
    nopr_mean_comment.append(i['commentCount'])
    nopr_mean_like.append(i['likeCount'])
    

print("熱門文章:")
print("留言人數:", sum(pr_mean_comment), " 平均留言人數:", round(sum(pr_mean_comment)/float(len(popular_data)),0))
print("按讚人數", sum(pr_mean_like), " 平均按讚人數:", round(sum(pr_mean_like)/float(len(popular_data)),0),'\n')

print("非熱門文章:")
print("留言人數:", sum(nopr_mean_comment), " 平均留言人數:", round(sum(nopr_mean_comment)/float(len(nopopular_data)),0))
print("按讚人數", sum(nopr_mean_like), " 平均按讚人數:", round(sum(nopr_mean_like)/float(len(nopopular_data)),0),'\n')
```
