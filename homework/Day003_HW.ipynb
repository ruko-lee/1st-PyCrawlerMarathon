### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異

```python
# File I/O
import os, sys

with open("./data/Day003_example/64_Week24_CH.xml", "r", encoding="utf-8") as fh:
    xml = fh.read()
print(xml)
```

```python
import xmltodict

#存取檔案 sample.xml
with open("./data/Day003_example/64_Week24_CH.xml", "r", encoding="utf-8") as fd:
    doc=dict(xmltodict.parse(fd.read()))

# 查看資料內容
print(doc)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

locations=doc['cwbopendata']['dataset']['locations']['location']

countL = 0
for location in locations:
    locationName = locations[countL]['locationName']
    countL +=1
    print(countL, locationName, end=', ')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

countL = 0
for location in locations:
    locationName = locations[countL]['locationName']
    FirstTimeS = locations[countL]['weatherElement'][0]['time'][0]['startTime']
    FirstTimeE = locations[countL]['weatherElement'][0]['time'][0]['endTime']
    FirstMeasure = locations[countL]['weatherElement'][0]['time'][0]['elementValue']['value']
    countL +=1
    print(countL, locationName)
    print('',FirstTimeS, FirstTimeE, FirstMeasure, sep='\t')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

locationName = locations[0]['locationName']
print(locationName)

for countL in locations[0]['weatherElement'][0]['time']:
    print(countL['startTime'], countL['endTime'], countL['elementValue']['measures'], countL['elementValue']['value'], sep='\t')
```
